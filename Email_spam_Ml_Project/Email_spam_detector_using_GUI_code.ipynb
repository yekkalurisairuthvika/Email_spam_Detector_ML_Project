```python
import tkinter as tk
from tkinter import Label, Button, Entry, messagebox
from tkinter import Label, Button, Entry
from PIL import Image, ImageTk
from tkinter import messagebox
import pandas as pd


spam_keywords = ["viagra", "lottery", "free money", "win", "prize", "claim", "urgent", "call", "free", "2", "txt", "text", "u", "ur", "mobil", "stop", "repli", "claim", "4", "get", "new", "servic", "tone", "send", "nokia", "contact", "avoid", "phone", "cash", "pleas", "wek", "c", "collect", "min"]

# Create the main window
main_window = tk.Tk()
main_window.title('Main Window')
main_window.geometry('800x600')


bg_image_main = Image.open("cyimg.png")  # Provide the path to your image
bg_image_main = ImageTk.PhotoImage(bg_image_main)

# Create a label to display the background image in the main window
bg_label_main = Label(main_window, image=bg_image_main)
bg_label_main.place(relwidth=1, relheight=1)


# Function to open a pop-up window
def open_popup():
    popup = tk.Toplevel()
    popup.title('SPAM Checker')
    popup.geometry('500x500')
    
    bg_image_popup = Image.open("cyberimg.png")  # Provide the path to your image
    bg_image_popup = ImageTk.PhotoImage(bg_image_popup)

    # Create a label to display the background image in the pop-up window
    bg_label_popup = Label(popup, image=bg_image_popup)
    bg_label_popup.place(relwidth=1, relheight=1)
    

    # Your existing code
    head = Label(popup, text='SPAM CHECKER', font=('helvetica', 24, 'bold'))
    head.pack()
    e = Entry(popup, width=500, borderwidth=6)
    e.pack()
    b = Button(popup, text='Check', font=('helvetica', 20, 'bold'), fg='white', bg='green', command=lambda: sms(e.get(), popup))
    b.pack()

    # Run the pop-up window
    popup.mainloop()

# Function for the SPAM Checker
def sms(message, popup):
    # Check if any spam keyword is present in the message
    for keyword in spam_keywords:
        if keyword in message.lower():
            result = "This message is SPAM."
            break
    else:
        result = "This message is not SPAM."

    # Display the result in the pop-up window
    result_label = Label(popup, text=result, font=('helvetica', 16))
    result_label.pack()

# Add an "Open" button in the main window
open_button = Button(main_window, text='Open', command=open_popup)
open_button.place(relx=0.5, rely=0.5, anchor='center')

# Run the main window
main_window.mainloop()
```
